## Cómo llegué aquí  

Como muchos en nuestro campo, mi camino hacia la analítica fue tortuoso. En la escuela, las matemáticas se convirtió en un tema que evité activamente; gran parte de ello parecía enteramente teórico. Hice algunos cursos en estadística y econometría que me llamaron la atención. Era un  soplo de aire fresco para aplicar las matemáticas a algún fin concreto.  

Es cierto que esta exposición a las estadísticas fue escasa. Asistí a una universidad de artes liberales, donde adquirí sólidas habilidades de escritura y pensamiento, pero pocas cuantitativas. Cuando yo cuando llegué a mi primer trabajo de tiempo completo, me sorprendió la profundidad y amplitud de los datos que tenía encargado de la gestión. Gran parte de estos datos se encontraban en hojas de cálculo y eran difíciles de obtener mucho valor sin una limpieza y preparación intensas.  
  
Parte de esta “disputa de datos” es de esperarse; el New York Times ha informado que los científicos de datos dedican entre el 50% y el 80% de su tiempo a preparar datos para su análisis. Pero gané se preguntó si había formas más eficientes de limpiar, administrar y almacenar datos. En particular, quería hacer esto para poder dedicar más tiempo a analizar los datos. Después de todo, siempre encontré el análisis estadístico algo aceptable: manual y hoja de cálculo propensa a errores. preparación de datos, no tanto.  

Como disfrutaba escribir (gracias, título en artes liberales), comencé a escribir un blog sobre consejos que aprendí en Excel. Gracias a la buena voluntad y al trabajo duro, el blog ganó fuerza y ​​le atribuyo gran parte de mi éxito profesional. Eres bienvenido a visítenos en stringfestanalytics.com; sigo publicando regularmente en Excel y análisis generalmente.  
  
Cuando comencé a aprender más sobre Excel, mi interés se extendió a otras herramientas de análisis y
técnicas. En este punto, los lenguajes de programación de código abierto R y Python habían ganado una popularidad significativa en el mundo de los datos. Pero mientras avanzaba al comprender estos lenguajes, sentí fricciones innecesarias en el camino de aprendizaje.

## Es un mal enfoque de instrucción.  

Los usuarios de Excel entienden intuitivamente cómo trabajar con datos: pueden ordenar, filtrar,
etc. Saben qué disposiciones facilitan el análisis y lo que significa mucha limpieza. Esta es una gran cantidad de conocimientos sobre los que basarse. La instrucción lo utilizará para cerrar la brecha entre las hojas de cálculo y la codificación. Desafortunadamente, la mayoría de las instrucciones queman el puente por desprecio. Las investigaciones indican que relacionar lo que has aprendido con lo que ya sabes es poderoso. Como lo expresaron Peter C. Brown en Make It Stick: La ciencia del éxito.  

## Aprendizaje (Prensa Belknap):  

Cuanto más puedas explicar sobre la forma en que tu nuevo aprendizaje se relaciona con tu aprendizaje anterior conocimiento, más fuerte será su comprensión del nuevo aprendizaje y más conexión tendrá con instrucciones que crees que te ayudarán a recordarlas más adelante.  

Como usuario de Excel, puede resultar difícil relacionar nuevas ideas con lo que ya sabe cuando
te dicen (erróneamente) que lo que ya sabes es basura. Este libro adopta un enfoque diferente y se basa en sus conocimientos previos sobre hojas de cálculo para que pueda tener un marco claro en mente a medida que avanzamos hacia R y Python.  

Tanto las hojas de cálculo como los lenguajes de programación son valiosas herramientas de analíticas; no es necesario abandonar Excel una vez que hayas aprendido R y Python.  

## Esta no es una lista de lavandería  

Una cosa que me encanta de la analítica es que casi siempre hay varias formas de hacer las la misma cosa. Es probable que te demuestre cómo hacer algo de una manera cuando estés familiarizado con otra. Mi objetivo en este libro es utilizar Excel como herramienta de enseñanza para el análisis y ayudar a los lectores transfieran este conocimiento a R y Python. Si tuviera que hacer de esto un volcado de cerebros todas las formas de completar una determinada tarea de análisis o limpieza de datos, el libro perdería su centrarse en este objetivo particular.  

Quizás prefieras formas alternativas de hacer algo; incluso puedo estar de acuerdo contigo en que,
dadas diferentes circunstancias, existe un mejor enfoque. Sin embargo, dadas las circunstancias y
posturas de este libro y sus objetivos, he decidido cubrir ciertas técnicas y excluir otras. Hacer lo contrario correría el riesgo de convertir el libro en un insulso manual de instrucciones, manual en lugar de una guía precisa para avanzar en el análisis.  

## No entrar en pánico  

Como autor, espero que me encuentren tranquilo y accesible. Sin embargo, tengo una regla para este libro: ¡que no cunda el pánico! Es cierto que aquí hay una curva de aprendizaje pronunciada, ya que explorará no solo la probabilidad y la estadística, sino también dos programas de programación.  

Este libro le presentará conceptos de estadística, informática, y más. Puede que al principio te resulten discordantes, pero empezarás a internalizarlos con el tiempo. Permítete aprender por prueba y error.  

Creo firmemente que con el conocimiento que tienes sobre Excel, este es un pedido alcanzable para un libro. Puede haber momentos de frustración y síndrome del impostor; Nos pasa a todos. No permita que estos momentos eclipsen el progreso real que logrará aquí.
